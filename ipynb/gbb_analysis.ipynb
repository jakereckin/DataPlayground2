```python
import pandas as pd, numpy as np
```

```python
game_stats = pd.read_csv('Basketball_2021_2022_Stats.csv')
```

```python
game_stats = game_stats.fillna(0)
game_stats = game_stats[game_stats['Field Goals Attempted'] > 0] 
game_stats = game_stats[game_stats['Field Goals Attempted_Allowed'] > 0]
game_stats['Offensive_Efficieny_Difference'] = game_stats['Offensive_Efficiency'] - game_stats['Offensive_Efficiency_Allowed']
game_stats.sort_values(by = 'Offensive_Efficieny_Difference').to_csv('test_for_analytics.csv', index = False)
```

```python
game_stats = game_stats[game_stats['Turnovers'] > 0]
game_stats = game_stats[game_stats['Turnovers_Allowed'] > 0]
game_stats.sort_values(by = 'Offensive_Efficieny_Difference')
```

```python
game_stats.to_csv('Basketball_Final_Stats.csv', index = False)
```
