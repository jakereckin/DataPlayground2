```python
game_stats_full['FGA'] = game_stats_full['2FGA'] + game_stats_full['3FGA']
game_stats_full['FGM'] = game_stats_full['2FGM'] + game_stats_full['3FGM']
game_stats_full['Points'] = (2 * game_stats_full['2FGM']) + (3 * game_stats_full['3FGM']) + game_stats_full['FTM']
game_stats_full
```

```python
def get_ppa_two(row):
    total_points = 2 * row['2FGM']
    total_attempts = row['2FGA']
    if total_attempts > 0:
        return total_points / total_attempts
    else:
        return 0

def get_ppa_three(row):
    total_points = 3 * row['3FGM']
    total_attempts = row['3FGA']
    if total_attempts > 0:
        return total_points / total_attempts
    else:
        return 0

def get_total_ppa(row):
    total_points = 2 * row['2FGM'] + 3 * row['3FGM']
    total_attempts = row['FGA']
    if total_attempts > 0:
        return total_points / total_attempts
    else:
        return 0

def team_scoring_poss(row):
    team_attempts = row['FGM']
    inner_part = (1 - (1 - (row['FTM'] / row['FTA']) ** 2))
    return team_attempts + inner_part *  row['FTA'] * 0.4

def team_play_percent(row):
    numerator = row['Scoring_Possesions']
    denom = row['FGA'] + row['FTA'] * 0.4 + row['Turnover']
    if denom > 0:
        return numerator / denom
    else:
        return 0

def possesions(row):
    fga = row['FGA']
    other_team_def_rebounds = (row['FGA'] - row['FGM']) - row['Off_Reb']
    reb_part = row['Off_Reb']/(row['Off_Reb'] + other_team_def_rebounds)
    fg_part = row['FGA'] - row['FGM']
    return fga - reb_part * fg_part * 1.07 + row['Turnover'] + row['FTA'] * 0.4

def won_game(row):
    if row['NDA_Score'] > row['Opponent_Score']:
        return 1
    else:
        return 0


def final_difference(row):
    return row['NDA_Score'] - row['Opponent_Score']

def offensive_efficiency(row):
    num = row['FGM'] + row['Assist']
    denom = row['FGA'] - row['Off_Reb'] + row['Assist'] + row['Turnover']
    if denom != 0:
        return num / denom
    else:
        return 0

def efficient_offense(row):
    inner = .76 * row['Assist'] + row['Points']
    return inner * row['Offensive_Efficiency']

def effective_fgp(row):
    num = row['FGM'] + (.5 * row['3FGM'])
    denom = row['FGA']
    if denom > 0:
        return num / denom
    else:
        return 0
```

```python
game_stats_full['Offensive_Efficiency'] = game_stats_full.apply(offensive_efficiency, axis = 1)
game_stats_full['Eff_Points'] = game_stats_full.apply(efficient_offense, axis = 1)
game_stats_full['EFG%'] = game_stats_full.apply(effective_fgp, axis = 1)
game_stats_full['2PPA'] = game_stats_full.apply(get_ppa_two, axis = 1)
game_stats_full['3PPA'] = game_stats_full.apply(get_ppa_three, axis = 1)
game_stats_full['PPA'] = game_stats_full.apply(get_total_ppa, axis = 1)
game_stats_full['Won_Flag'] = game_stats_full.apply(won_game, axis = 'columns')
game_stats_full['Score_Diff'] = game_stats_full.apply(final_difference, axis = 'columns')
game_stats_full['Scoring_Possesions'] = game_stats_full.apply(team_scoring_poss, axis = 'columns')
game_stats_full['Possesions'] = game_stats_full.apply(possesions, axis = 'columns')
game_stats_full['Team_Play%'] = game_stats_full.apply(team_play_percent, axis = 'columns')
game_stats_full
```

```python
from pyomo.environ import*
import pandas as pd
players = ['HB', 'TF', 'GG', 'SW', 'KW', 'PM', 'PD', 'HG', 'KR', 'MG', 'DW']
off_eff = [.72 * .9, .55 * 1.1, .67 * .8, .73, .88 * .7, 
           .57 * 1.1, .38, .7 * .8, .2, .3, .2]
min_need = [15, 30, 30, 15, 5, 10, 5, 5, 0, 0, 0]
off_dict = dict(zip(players, off_eff))
min_dict = dict(zip(players, min_need))

model = ConcreteModel()

model.players = Var(players, domain = NonNegativeReals)

model.totals = Objective(expr = sum(off_dict[player] * model.players[player] for player in players), sense = maximize)

model.minutes = ConstraintList()
for player in players:
    model.minutes.add(model.players[player] >= min_dict[player])

model.max_mins = ConstraintList()
for player in players:
    model.max_mins.add(model.players[player] <= 36)

model.total_mins = ConstraintList()
model.total_mins.add(sum(model.players[player] for player in players) == 180)

solver = SolverFactory('glpk')
solver.solve(model)

dvars = pd.DataFrame([model.players[player]() for player in players],
                     index = players,
                     columns = ['Minutes'])
dvars
```
