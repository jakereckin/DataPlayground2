```python
my_excel_file = r'C:\Users\Jake\Documents\Grad School\DS 735\VTL Strength Index - External Indicator Data - 12JULY2022.xlsx'
data = pd.read_excel(my_excel_file).drop(columns = ['Date'])
data
```

```python
def get_redundanent_pairs(df):
    pairs_to_drop = set()
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0, i + 1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

def get_correlations(df):
    au_corr = df.corr().unstack()
    labels_to_drop = get_redundanent_pairs(df)
    au_corr_final = au_corr.drop(labels = labels_to_drop).sort_values(ascending = True)
    return au_corr_final
```

```python
new_frame = get_correlations(data)
#new_frame_strength = new_frame[(new_frame['Feature 1'] == 'VTL Strength') | (new_frame['Feature 2'] == 'VTL Strength')]
#new_frame_strength
my_data = new_frame['VTL Strength']
```

```python
from matplotlib.colors import TwoSlopeNorm
fig, ax = plt.subplots(figsize = (15, 10))
norm = TwoSlopeNorm(vmin = -1, vcenter = 0, vmax = 1)
colors = [plt.cm.RdYlBu_r(norm(c)) for c in my_data.values]
my_data.plot.barh(color = colors)
```
